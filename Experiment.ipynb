```python
from parser import Parser
from demoparser2 import DemoParser
```

```python
parser: DemoParser = DemoParser("heroic-vs-3dmax-m1-dust2.dem")
parser_attributes = ["total_rounds_played"]
bomb_defused = parser.parse_event(
    "bomb_defused", other=parser_attributes
)
bomb_planted = parser.parse_event(
    "bomb_planted", other=parser_attributes
)
```

```python
round = bomb_defused['total_rounds_played'].iloc[0]
end_t = bomb_defused[bomb_defused['total_rounds_played'] == round].iloc[0]['tick']
start_t = bomb_planted[bomb_planted['total_rounds_played'] == round].iloc[0]['tick']
```

```python
ATTRIBUTES = ["team_name", "is_alive"]
TRACE_ATTRIBUTES = ["last_place_name", "tick"]

df = parser.parse_ticks(
    ATTRIBUTES + TRACE_ATTRIBUTES,
    ticks=range(int(start_t), int(end_t + 1)),
)
df
```
